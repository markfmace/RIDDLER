```python
import numpy as np
import matplotlib.pyplot as plt 
import random
import matplotlib.mlab

def battle(n1,n2):
    living=n1
    dead=n2
    surviving=True
    while(surviving==True):
        fight=random.uniform(0,1)
        if(fight<0.5):
            living=living
            dead=dead-1
        else:
            living=living-1
            dead=dead+1
    
        if(living<1 or dead<1):
            surviving = False

    return([living,dead])



# print(battle(5,5))
result2=[]

for j in range(1,450,10):
    for k in range(1,20,1):
        result=[]
        numsamples=1000
        for i in range(0,numsamples):
            result.append(battle(j,k))

        result=np.array(result)
#         print(result)

        result2.append([j,k,np.sum(1 for x in result[:,0] if x!=0)/numsamples])
  

result2=np.array(result2)

print(np.shape(result2))
x=result2[:,0]
y=result2[:,1]
z=result2[:,2]
print(max(z))
plt.tricontourf(x,y,z)
plt.legend(frameon=False)
plt.xlabel('Living arym size')
plt.ylabel('Dead arym size')
# plt.clabel('Survival percentage')

plt.colorbar()
# print(result2)

# ax.plot_surface(result2[:,0],result2[:,1],result2[:,2])  
# 
# result3<-NULL
# for (i in unique(result2$living)){
#   half<-spline(x = result2[result2$living==i,]$livewin, y = result2[result2$living==i,]$dead, xout=50)$y
#   result3<-rbind(result3, c(living=i,dead=half))
# }
# result3<-as.data.frame(result3)
# ggplot(result3, aes(x=sqrt(living), y=dead))+geom_line()
```
